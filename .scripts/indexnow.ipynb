```python
import os
import sys
import re

def remove_first_pg_tag(folder):
    pattern = re.compile(r'<!--pg-->')

    for root, dirs, files in os.walk(folder):
        for file in files:
            file_path = os.path.join(root, file)

            # Chỉ làm với các file text, bạn có thể thêm điều kiện lọc file type ở đây nếu cần
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()

                # Nếu có match thì xóa match đầu tiên
                if '<!--pg-->' in content:
                    new_content = pattern.sub('', content, count=1)

                    with open(file_path, 'w', encoding='utf-8') as f:
                        f.write(new_content)

                    print(f'Removed in: {file_path}')
            except Exception as e:
                print(f'Error processing {file_path}: {e}')
remove_first_pg_tag ('../.docsource/kinh-tang-chi/sujato/2')
```

```python
from pathlib import Path

INPUT_DIR = "../docs/kinhtruongbo/pali/"
OUTPUT_DIR = Path("output/mn/")

OUTPUT_DIR.mkdir(exist_ok=True)

# lấy danh sách file .md và sort theo tên
files = sorted(Path(INPUT_DIR).glob("*.md"))

# merge mỗi 2 file
for i in range(0, len(files), 2):
    chunk = files[i:i+2]

    merged = []

    for file in chunk:
        content = file.read_text(encoding="utf-8")
        merged.append(content)

    output_content = "\n\n".join(merged)

    output_name = f"dn-{(i//2)+1:02d}.md"
    output_path = OUTPUT_DIR / output_name

    output_path.write_text(output_content, encoding="utf-8")

    print(f"Created: {output_path}")
```
