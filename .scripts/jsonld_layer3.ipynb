```python
import json
import os
import re
from pathlib import Path

BASE_URL = "https://kinhnikaya.org"

def build_has_part(folder_path):

    # input: folder path
    file_path = Path(folder_path)

    # ưu tiên: mucluc.md -> meta/mucluc.md
    mucluc_1 = file_path / "mucluc.md"
    mucluc_2 = file_path / "meta" / "mucluc.md"

    if mucluc_1.exists():
        md_file = mucluc_1
    elif mucluc_2.exists():
        md_file = mucluc_2
    else:
        raise FileNotFoundError(
            f"Không tìm thấy '{mucluc_1}' hoặc '{mucluc_2}'"
        )


    text = Path(md_file).read_text(encoding="utf-8")

    # lấy tất cả markdown links
    links = re.findall(r'\[.*?\]\((.*?)\)', text)

    result = []

    for link in links:
        path = link.strip()

        # bỏ .md ở cuối, trừ intro/outro
        if "/intro/" in path or "/outro/" in path:
            suffix = "#article"
            final_path = path
        else:
            suffix = "#chapter"
            final_path = re.sub(r'\.md$', '', path)

        result.append(
            { "@id": f'{BASE_URL}{final_path}{suffix}'}
        )
    return result


def build_layer3_jsonld(kinh: str, author: str, has_part: list | None = None) -> dict:
    """
    Sinh JSON-LD tầng 3 cho trang bản dịch.

    Parameters
    ----------
    kinh    : key bộ kinh trong CONFIG, vd 'kinhtruongbo'
    author  : key bản dịch,            vd 'thichminhchau' hoặc 'sujato-vi'
    has_part: list các dict {"@id": "...#chapter"} — để trống thì bỏ qua field hasPart
    """
    bộ    = CONFIG[kinh]
    ver   = bộ["versions"][author]
    slug  = author                        # vd: 'thichminhchau'
    label = bộ["label"]                   # vd: 'Kinh Trường Bộ'

    page_url    = f"{BASE}/{kinh}/{slug}/"
    book_id     = f"{page_url}#book"
    parent_url  = f"{BASE}/{kinh}/"
    abstract_id = f"{parent_url}#book"

    # ── translator: hỗ trợ 1 người (dict) hoặc nhiều người (list) ──
    raw_tr = ver["translator"]
    if isinstance(raw_tr, list):
        translator_field = [
            {"@type": "Person", **t} for t in raw_tr
        ]
    else:
        translator_field = {"@type": "Person", **raw_tr}

    # ── Book node ──────────────────────────────────────────────────
    book = {
        "@type": "Book",
        "@id":   book_id,
        "mainEntityOfPage": {
            "@type": "WebPage",
            "@id":   page_url
        },
        "name":        ver["name"],
        "bookEdition": ver["bookEdition"],
        "url":         page_url,
        "inLanguage":  ver["inLanguage"],
        "isAccessibleForFree": True,
        "publisher":   {"@id": f"{BASE}/#org"},
        "image":       f"{BASE}/covers/{ver['image']}",
        "author": {
            "@type":  "Person",
            "name":   "Gotama Buddha",
            "sameAs": "https://en.wikipedia.org/wiki/Gautama_Buddha"
        },
        "translator":        translator_field,
        "translationOfWork": {"@id": abstract_id}
    }


    if has_part:
        book["hasPart"] = has_part
    else:
        index_file = f"../docs/{kinh}/{author}"
        book["hasPart"] =  build_has_part(index_file)


    # ── BreadcrumbList ─────────────────────────────────────────────
    breadcrumb = {
        "@type": "BreadcrumbList",
        "itemListElement": [
            {"@type": "ListItem", "position": 1, "name": "Trang chủ",  "item": f"{BASE}/"},
            {"@type": "ListItem", "position": 2, "name": label,         "item": parent_url},
            {"@type": "ListItem", "position": 3, "name": ver["breadcrumb3"], "item": page_url}
        ]
    }

    return {
        "@context": "https://schema.org",
        "@graph": [book, breadcrumb]
    }


# ── CHẠY ──────────────────────────────────────────────────────────────────────
kinh   = "kinhtangchi"
author = "sujato-vi"   # đổi thành 'sujato-vi', 'sujato-en', …
#has_part = [{}] # tu build has_part

result = build_layer3_jsonld(kinh, author)

# JSON-LD pretty format
jsonld_result = json.dumps(result, ensure_ascii=False, indent=2)

# dữ liệu từ CONFIG
bộ   = CONFIG[kinh]
ver  = bộ["versions"][author]

label        = bộ["label"]
breadcrumb3  = ver["breadcrumb3"]
image        = ver["image"]
name         = ver["name"]
markdown_output = f"""---
title: {label} - {name}
head:
  - - script
    - type: application/ld+json
    - |
       {jsonld_result.replace(chr(10), chr(10) + '        ')}

---

<h1 style="text-align: center;">{label}</h1>
<h2 style="text-align: center;">{name}</h2>
<br>
<a href="/{kinh}/{author}/-- mục đầu tiên">
<figure style="text-align: center;">
    <img src="/covers/{image}" style="max-width:400px;width:100%;display:block;margin:0 auto;" alt="{label} {name}">
</figure>
</a>

<br/>
<hr/>
<br/>
<h2 style="text-align: center;">Mục Lục</h2>
<br/>

<!--@include: ./meta/mucluc.md-->
"""

print(markdown_output)

# print(json.dumps(result, ensure_ascii=False, indent=2))
```
