```python
# validate config
import re
from collections import defaultdict

BASE = "https://kinhnikaya.org"


def validate_config(CONFIG):
    errors = []
    warnings = []

    # ==========================================================
    # build index các version để kiểm tra @id có tồn tại hay không
    # ==========================================================

    valid_book_ids = set()
    valid_work_ids = set()

    for collection in CONFIG:
        valid_work_ids.add(f"{BASE}/{collection}/#work")

        for version in CONFIG[collection]["versions"]:
            valid_book_ids.add(f"{BASE}/{collection}/{version}/#book")

    # ==========================================================
    # validate từng collection
    # ==========================================================

    for collection, data in CONFIG.items():

        versions = data["versions"]

        breadcrumb_set = set()

        for version_name, version in versions.items():

            prefix = f"[{collection}/{version_name}]"

            # --------------------------------------------------
            # bắt buộc
            # --------------------------------------------------

            required = [
                "name",
                "bookEdition",
                "image",
                "inLanguage",
                "genre",
                "breadcrumb3",
            ]

            for field in required:
                if field not in version:
                    errors.append(f"{prefix} missing '{field}'")

            # --------------------------------------------------
            # image
            # --------------------------------------------------

            image = version.get("image")

            if image:

                if not image.startswith(collection):
                    errors.append(
                        f"{prefix} image '{image}' phải bắt đầu bằng '{collection}'"
                    )

            # --------------------------------------------------
            # breadcrumb3 unique
            # --------------------------------------------------

            bc = version.get("breadcrumb3")

            if bc:

                if bc in breadcrumb_set:
                    errors.append(
                        f"{prefix} breadcrumb3 bị trùng '{bc}'"
                    )

                breadcrumb_set.add(bc)

            # --------------------------------------------------
            # translationOfWork
            # --------------------------------------------------

            tow = version.get("translationOfWork")

            if version.get("genre") == "Translation":

                if tow is None:
                    errors.append(
                        f"{prefix} Translation nhưng thiếu translationOfWork"
                    )

                else:

                    expected = f"{BASE}/{collection}/#work"

                    actual = tow.get("@id")

                    if actual != expected:
                        errors.append(
                            f"{prefix} translationOfWork sai\n"
                            f"    expected : {expected}\n"
                            f"    actual   : {actual}"
                        )

            # --------------------------------------------------
            # Root text không nên có translationOfWork
            # --------------------------------------------------

            if version.get("genre") == "Root text":

                if "translationOfWork" in version:
                    warnings.append(
                        f"{prefix} Root text không cần translationOfWork"
                    )

            # --------------------------------------------------
            # isBasedOn
            # --------------------------------------------------

            if "isBasedOn" in version:

                refs = version["isBasedOn"]

                if isinstance(refs, dict):
                    refs = [refs]

                for ref in refs:

                    if "@id" not in ref:
                        errors.append(f"{prefix} isBasedOn thiếu @id")
                        continue

                    target = ref["@id"]

                    # ---------- tồn tại ----------

                    if target not in valid_book_ids:
                        errors.append(
                            f"{prefix} isBasedOn trỏ tới @id không tồn tại\n"
                            f"    {target}"
                        )

                    # ---------- đúng collection ----------

                    m = re.match(
                        rf"^{re.escape(BASE)}/([^/]+)/",
                        target
                    )

                    if m:

                        target_collection = m.group(1)

                        if target_collection != collection:

                            errors.append(
                                f"{prefix} isBasedOn trỏ sai collection\n"
                                f"    current : {collection}\n"
                                f"    target  : {target_collection}"
                            )

            # --------------------------------------------------
            # translator
            # --------------------------------------------------

            if "translator" in version:

                tr = version["translator"]

                if isinstance(tr, dict):
                    tr = [tr]

                elif not isinstance(tr, list):
                    errors.append(
                        f"{prefix} translator phải là dict hoặc list"
                    )
                    tr = []

                for t in tr:

                    if not isinstance(t, dict):
                        errors.append(
                            f"{prefix} translator có phần tử không phải dict"
                        )
                        continue

                    if "name" not in t:
                        errors.append(
                            f"{prefix} translator thiếu name"
                        )

            # --------------------------------------------------
            # inLanguage
            # --------------------------------------------------

            lang = version.get("inLanguage")

            if lang not in ("vi", "en", "pi"):
                warnings.append(
                    f"{prefix} inLanguage '{lang}' lạ"
                )

            # --------------------------------------------------
            # image extension
            # --------------------------------------------------

            if image:

                if not image.endswith(".webp"):
                    warnings.append(
                        f"{prefix} image nên dùng webp"
                    )

            # --------------------------------------------------
            # Translation nên có description
            # --------------------------------------------------

            if version.get("genre") == "Translation":

                if "description" not in version:
                    warnings.append(
                        f"{prefix} thiếu description"
                    )

            # --------------------------------------------------
            # Root text nên có description
            # --------------------------------------------------

            if version.get("genre") == "Root text":

                if "description" not in version:
                    warnings.append(
                        f"{prefix} Root text thiếu description"
                    )

    # ==========================================================
    # report
    # ==========================================================

    print("=" * 70)

    if errors:

        print(f"❌ {len(errors)} ERROR(S)\n")

        for e in errors:
            print(e)

    else:
        print("✅ Không có ERROR")

    print()

    if warnings:

        print(f"⚠️  {len(warnings)} WARNING(S)\n")

        for w in warnings:
            print(w)

    else:
        print("✅ Không có WARNING")

    print("=" * 70)

    return errors, warnings

errors, warnings = validate_config(CONFIG)
print (errors, warnings)
```

```python
import json
import os
import re
from pathlib import Path

BASE_URL = "https://kinhnikaya.org"

def build_has_part(folder_path):

    # input: folder path
    file_path = Path(folder_path)

    # ưu tiên: mucluc.md -> meta/mucluc.md
    mucluc_1 = file_path / "mucluc.md"
    mucluc_2 = file_path / "meta" / "mucluc.md"

    if mucluc_1.exists():
        md_file = mucluc_1
    elif mucluc_2.exists():
        md_file = mucluc_2
    else:
        raise FileNotFoundError(
            f"Không tìm thấy '{mucluc_1}' hoặc '{mucluc_2}'"
        )


    text = Path(md_file).read_text(encoding="utf-8")

    # lấy tất cả markdown links
    links = re.findall(r'\[.*?\]\((.*?)\)', text)

    result = []
    first_link = None
    for link in links:
        path = link.strip()

        # bỏ .md ở cuối, trừ intro/outro
        if "/intro/" in path or "/outro/" in path:
            suffix = "#article"
            final_path = path
        else:
            suffix = "#chapter"
            final_path = re.sub(r'\.md$', '', path)

        if not first_link:
            first_link = f'{path}'

        result.append(
            { "@id": f'{BASE_URL}{final_path}{suffix}'}
        )
    return result, first_link

def build_layer3_jsonld(kinh: str, author: str, has_part: list | None = None) -> dict:
    """
    Sinh JSON-LD tầng 3 cho trang bản dịch.
    """

    bộ = CONFIG[kinh]
    ver = bộ["versions"][author]
    slug = author
    label = bộ["label"]

    page_url = f"{BASE}/{kinh}/{slug}/"

    webpage_id = f"{page_url}#webpage"
    book_id = f"{page_url}#book"
    breadcrumb_id = f"{page_url}#breadcrumb"

    parent_url = f"{BASE}/{kinh}/"
    work_id = f"{parent_url}#work"

    # ── translator ────────────────────────────────────────────────
    translator_field = None
    if ver.get("translator"):
        raw_tr = ver["translator"]
        if isinstance(raw_tr, list):
            translator_field = [
                {"@type": "Person", **t} for t in raw_tr
            ]
        else:
            translator_field = {
                "@type": "Person",
                **raw_tr
            }

    # ── WebPage ───────────────────────────────────────────────────
    webpage = {
        "@type": "WebPage",
        "@id": webpage_id,
        "url": page_url,
        "mainEntity": {
            "@id": book_id
        },
        "isPartOf": {
            "@id": f"{BASE}/#website"
        },
        "breadcrumb": {
            "@id": breadcrumb_id
        }
    }

    # ── Book ──────────────────────────────────────────────────────
    book = {
        "@type": "Book",
        "@id": book_id,
        "name": ver["name"],
        "description": ver.get("description"),          # TODO: thêm description vào CONFIG
        "bookEdition": ver["bookEdition"],
        "url": page_url,
        "inLanguage": ver["inLanguage"],
        "isAccessibleForFree": True,
        "publisher": {"@id": f"{BASE}/#org"},
        "image": f"{BASE}/covers/{ver['image']}",
        "translationOfWork": {"@id": work_id },
        "genre": ver["genre"]
    }
    
    if ver.get("isBasedOn"):
      book["isBasedOn"]= ver.get("isBasedOn")
        

    if translator_field:
        book["translator"] = translator_field

    # hasPart
    firstLink = None
    if has_part:
        book["hasPart"] = has_part
    else:
        index_file = f"../../docs/{kinh}/{author}"
        hasPartb, firstLink = build_has_part(index_file)
        book["hasPart"] = hasPartb
    
    if firstLink:
      book["potentialAction"] = {
        "@type": "ReadAction",
        "target": {
          "@type": "EntryPoint",
          "urlTemplate": firstLink,
        }
      }
        # book["potentialAction"]  = 


    # ── Breadcrumb ────────────────────────────────────────────────
    breadcrumb = {
        "@type": "BreadcrumbList",
        "@id": breadcrumb_id,
        "itemListElement": [
            {
                "@type": "ListItem",
                "position": 1,
                "name": "Trang chủ",
                "item": f"{BASE}/"
            },
            {
                "@type": "ListItem",
                "position": 2,
                "name": label,
                "item": parent_url
            },
            {
                "@type": "ListItem",
                "position": 3,
                "name": ver["breadcrumb3"],
                "item": page_url
            }
        ]
    }

    jsonld_result = {
        "@context": "https://schema.org",
        "@graph": [
            webpage,
            book,
            breadcrumb
        ]
    }

    return jsonld_result, firstLink

# ── CHẠY ──────────────────────────────────────────────────────────────────────
kinh   = "kinhtuongung"
author = "c-sujato-tmc-vi"   # đổi thành 
author = 'sujato-vi'
author = 'sujato-en'
author = 'pali'
#has_part = [{}] # tu build has_part

result, firstLink = build_layer3_jsonld(kinh, author)

# JSON-LD pretty format
jsonld_result = json.dumps(result, ensure_ascii=False, indent=2)

# dữ liệu từ CONFIG
bộ   = CONFIG[kinh]
ver  = bộ["versions"][author]

label        = bộ["label"]
breadcrumb3  = ver["breadcrumb3"]
image        = ver["image"]
name         = ver["name"]
bookEdition  = ver["bookEdition"]
desc         = ver["description"]
markdown_output = f"""---
title: {label} - {name}
description: {desc} - {name} - {bookEdition}
head:
  - - script
    - type: application/ld+json
    - |
       {jsonld_result.replace(chr(10), chr(10) + '        ')}

---

<h1 style="text-align: center;">{label}</h1>
<h2 style="text-align: center;">{name}</h2>
<br>
<a href= "{firstLink}">
<figure style="text-align: center;">
    <img src="/covers/{image}" style="max-width:400px;width:100%;display:block;margin:0 auto;" alt="{desc}">
</figure>
</a>

<br/>
<hr/>
<br/>
<h2 style="text-align: center;">Mục Lục</h2>
<br/>

<!--@include: ./meta/mucluc.md-->
"""

print(markdown_output)


```
