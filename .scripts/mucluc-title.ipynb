```python
from pathlib import Path

# Thư mục gốc cần quét
ROOT = Path("../docs")   # đổi thành Path("/đường/dẫn") nếu cần

HEADER = "---\ntitle: Mục lục\n---\n\n"

updated = []
skipped = []
errors = []

for file in ROOT.rglob("mucluc.md"):
    try:
        text = file.read_text(encoding="utf-8")

        # File đã có frontmatter
        if text.startswith("---"):
            skipped.append(file)
            print(f"[SKIP] {file}")
            continue

        file.write_text(HEADER + text, encoding="utf-8")
        updated.append(file)
        print(f"[UPDATE] {file}")

    except Exception as e:
        errors.append((file, str(e)))
        print(f"[ERROR] {file}: {e}")

print("\n========== SUMMARY ==========")
print(f"Updated : {len(updated)}")
print(f"Skipped : {len(skipped)}")
print(f"Errors  : {len(errors)}")

if skipped:
    print("\nFiles already containing frontmatter:")
    for f in skipped:
        print(" -", f)

if errors:
    print("\nErrors:")
    for f, e in errors:
        print(f" - {f}: {e}")
```
