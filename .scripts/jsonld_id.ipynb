```python
import os

BASE_URL = "https://kinhnikaya.org"
DOCS_ROOT = os.path.join(os.path.dirname(os.getcwd()), "docs")  # adjust if needed

folders = [
    "docs/kinhtruongbo/thichminhchau",
    "docs/kinhtruongbo/pali-vi",
]

results = []

for folder in folders:
    abs_folder = os.path.join(os.path.dirname(os.getcwd()), folder)
    if not os.path.isdir(abs_folder):
        print(f"[skip] not found: {abs_folder}")
        continue
    for fname in sorted(os.listdir(abs_folder)):
        if not fname.endswith(".md"):
            continue
        # strip docs/ prefix and .md suffix, build URL path
        rel = os.path.join(folder, fname)          # e.g. docs/kinhtruongbo/.../dn-001.md
        url_path = rel.removeprefix("docs/").removesuffix(".md")  # kinhtruongbo/.../dn-001
        url = f"{BASE_URL}/{url_path}#chapter"
        results.append(f'{{ "@id": "{url}" }}')

print("\n".join(results))
```
