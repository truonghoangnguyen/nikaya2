```python
import os

"""
Co note:
1. a.md -> a.p0.md + a.note.md
1. dich a.p0.md -> a.vi.p0.md (compare)
2. dich a.note.md -> a.vi.note.md (compare)
3. merge a.vi.p0.md + a.vi.note.md -> a.vi.md (nomal)


Quy trình như sau, từ file a.md
- hiển thị 2 nơi: 1) compate, 2) bình thường
- normal node:
 1. dùng footnote
 2. tất cả nội dung vào 1 file
- compare mode:
 1. không dùng được footnote
 2. phải chia ra note file riêng
 3. từ compare tạo link tới normal mode

Ta có 4 trường hợp từ 1 file markdown: a.md
1. không part, không note -> ghi a.md
2. không part, có note -> ghi a.md+a.note.md
3. có part, không note -> ghi a.p1.md, a.p2.md
4. có part, có note -> ghi a.p1.md, a.p2.md + a.note.md

Thực hiện, để hiển thị trong compare
Không part, không note file không cần làm gì (a.md)
Nếu file có part/có note thì chia part
 - a.md
 - a.p0.md
 - a.p0.md + a.note.md
 - a.p1.md, a.p2.md
 - a.p1.md, a.p2.md + a.note.md


"""

def split_markdown_file(filepath, split_points):
    """
    Splits a markdown file into multiple parts based on specified line numbers.

    Args:
        filepath: The path to the markdown file.
        split_points: A list of integers representing the line numbers
                      where the file should be split.  The line numbers are
                      *inclusive* for the *end* of each section. (1-based indexing).
    """

    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            lines = f.readlines()
    except FileNotFoundError:
        print(f"Error: File not found: {filepath}")
        return
    except Exception as e:
        print(f"Error reading file: {e}")
        return

    total_lines = len(lines)

    # Validate split points
    if not all(0 < point <= total_lines for point in split_points):
        print("Error: Split points must be within the range of lines in the file (1-based indexing).")
        return

    split_points = sorted(split_points)  # Ensure split points are in ascending order

    # Add 0 and total_lines for slicing convenience
    all_split_points = [0] + split_points + [total_lines]


    base_filename, ext = os.path.splitext(filepath)

    for i in range(len(all_split_points) - 1):
        start_line = all_split_points[i]
        end_line = all_split_points[i+1]

        part_lines = lines[start_line:end_line]

        output_filename = f"{base_filename}.p{i+1}{ext}"
        try:
            with open(output_filename, 'w', encoding='utf-8') as outfile:
                outfile.writelines(part_lines)
            print(f"Created part {i+1}: {output_filename} (lines {start_line + 1}-{end_line})")
        except Exception as e:
            print(f"Error writing to {output_filename}: {e}")


filepath= "../kinhtrungbo/nanamoli-bodhi-vi/001-the-root-of-all-things.vi.p0.md"
# filepath= "../kinhtrungbo/thichminhchau/001-kinh-phap-mon-can-ban.md"

split_markdown_file(filepath, [216, 239])
```

```python
def fix_text_string(text):
    lines = text.split('\n')

    # Tạo một danh sách mới để lưu các dòng sau khi xử lý
    fixed_lines = []

    i = 0
    while i < len(lines):
        current_line = lines[i].strip()

        # Kiểm tra nếu dòng hiện tại trống và dòng tiếp theo không trống
        if i < len(lines) - 1 and current_line == "" and lines[i+1].strip() != "":
            next_line = lines[i+1].strip()

            # Kiểm tra nếu dòng trước không trống và không kết thúc bằng dấu chấm hoặc dấu câu kết thúc câu
            if i > 0 and lines[i-1].strip() != "" and not lines[i-1].strip().endswith((".", "?", "!", '."', '?"', '!"')):
                # Kiểm tra nếu dòng tiếp theo bắt đầu bằng chữ thường hoặc dấu ngoặc vuông
                if (next_line[0].islower() or next_line.startswith("[")) and not next_line.startswith('"'):
                    # Bỏ qua dòng trống này
                    i += 1
                    continue

        # Thêm dòng hiện tại vào danh sách đã sửa
        fixed_lines.append(lines[i])
        i += 1

    return '\n'.join(fixed_lines)

# Test với đoạn text BAD
bad_text = """23. “He perceives unity as unity.[18] Having perceived
unity as unity, he conceives [himself as] unity, he conceives

[himself] in unity, he conceives [himself apart] from unity, he
conceives unity to be ‘mine,’ he delights in unity. Why is that?
Because he has not fully understood it, I say.

24. “He perceives diversity as diversity. Having perceived
diversity as diversity, he conceives [himself as] diversity, he
conceives [himself] in diversity, he conceives [himself apart]
from diversity, he conceives diversity to be ‘mine,’ he delights
in diversity. Why is that? Because he has not fully understood
it, I say."""

# Chạy hàm và in kết quả
fixed_text = fix_text_string(bad_text)
print(fixed_text)
```
